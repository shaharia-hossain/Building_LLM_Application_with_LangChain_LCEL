```python
#  load groq api key
api_key = os.getenv("GROQ_API")
if not api_key:
  raise EnvironmentError("critical Error: GROQ_API key is missing")
```

### architectural design 

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
# initial model
llm = ChatGroq(
  model="qwen/qwen3-32b",
  api_key = api_key, # provide api key here
  temperature = 0.7, # balance between randomness and accurecy
  max_retries = 2 # try 2 times if network or have any issues
)
```

```python
llm
```

```python
system_instruction = (
    "You are a professional AI Assistant. Your task is to answer the user's question "
    "accurately in {language}. Provide only the direct answer and do not show your internal thinking."
)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", system_instruction),
    ("human", "{user_query}")
])

parser = StrOutputParser() # this is for parse the output for getting clen data
```

### Phase 3

```python
#  create chain uign "|"
# pattern = input -> prompt -> model -> output
ai_chain = prompt | llm | parser
```

```python
response = ai_chain.invoke({
    "language": "Bangla",
    "user_query": "my name is shahria, im a devloper"
})
```

```python
print(f"Response: \n {response}")
```
